## Normality tests

```python
#converting the numeric data to a numpy array
pre_molt_data = df['Pre-molt'].to_numpy()
post_molt_data = df['Post-molt'].to_numpy()
```

Anderson-Darling Test

```python
pre_molt_result = stats.anderson(pre_molt_data)
post_molt_result = stats.anderson(post_molt_data)

print("Statistic for pre molt values:", pre_molt_result.statistic)
print("Statistic for post molt values:", post_molt_result.statistic)
print("Critical Values for pre molt values:", pre_molt_result.critical_values)
print("Critical Values for post molt values:", post_molt_result.critical_values)
print("Significance Levels pre molt values:", pre_molt_result.significance_level)
print("Significance Levels post molt values:", post_molt_result.significance_level)
```

Kolmogorov-Smirnov Test:

```python
pre_molt_result = stats.kstest(pre_molt_data, 'norm')
post_molt_result = stats.kstest(post_molt_data, 'norm')
print("KS Statistic for pre molt data:", pre_molt_result.statistic)
print("KS Statistic for post molt data:", post_molt_result.statistic)
print("P-value for pre molt data:", pre_molt_result.pvalue)
print("P-value for post molt data:", post_molt_result.pvalue)
```

Cramér-von Mises Test:

```python
pre_molt_result = sm.stats.diagnostic.normal_ad(pre_molt_data)
post_molt_result = sm.stats.diagnostic.normal_ad(post_molt_data)

pre_molt_statistic, pre_molt_pvalue = pre_molt_result
post_molt_statistic, post_molt_pvalue = post_molt_result

print("Test Statistic for pre molt data:", pre_molt_statistic)
print("Test Statistic for post molt data:", post_molt_statistic)
print("P-value for pre molt data:", pre_molt_pvalue)
print("P-value for post molt data:", post_molt_pvalue)
```

Shapiro-Wilk Test:

```python
pre_molt_statistic, pre_molt_p_value = stats.shapiro(pre_molt_data)
post_molt_statistic, post_molt_p_value = stats.shapiro(post_molt_data)
print("Test Statistic for pre molt data: ", pre_molt_statistic)
print("Test Statistic for post molt data: ", post_molt_statistic)
print("P-value for pre molt data: ", pre_molt_p_value)
print("P-value for post molt data: ", post_molt_p_value)
```

# Linear Least Squares Model

## Linear Model

```python
x=df["Post-molt"].to_numpy()
y=df["Pre-molt"].to_numpy()

x=x.reshape(-1, 1)
model = LinearRegression().fit(x,y)
r_sq = model.score(x,y)
print("Coeff of Determination: ",r_sq)
```

```python
b = model.intercept_
m = model.coef_

plt.plot(x, y, '.')
# add linear regression line to scatterplot
plt.plot(x, m*x+b)
plt.xlabel("Post molt")
plt.ylabel("Pre molt")
plt.title("Linear Model")
```

## Residuals

```python
y_pred = model.predict(x)

# Calculate residuals
residuals = y - y_pred
rs = pd.DataFrame(residuals, columns=['Residuals'])

rs.describe()
```

```python
print("Skewness\n",rs.skew())

print("Kurtosis\n",rs.kurtosis())
```

```python
# Create a smooth histogram using seaborn
sns.histplot(data=rs, kde=True, bins=20, common_norm=False, element="step", fill=True)

# Set plot labels and title
plt.xlabel("Size")
plt.ylabel("Density")
plt.title("Smooth Histogram of Residuals")

# Show the plot
plt.show()
```

```python
plt.axhline(0, color='black', linestyle='--', label='Zero Residual Line')

# Plotting residuals as points
plt.plot(residuals, '.', label='Residuals')

# Connecting each point to the zero-residual line
for i, res in enumerate(residuals):
    plt.plot([i, i], [0, res], color='gray', linestyle='--', linewidth=0.1)

# Customize the plot
plt.xlabel('Data Point')
plt.ylabel('Residual')
plt.title('Residuals Comparison')
plt.legend()
plt.show()
```

## QQ Plot

### Breusch–Pagan test, White test

```python
x = df["Post-molt"].to_numpy().reshape(-1, 1)
y = df["Pre-molt"].to_numpy()

# Add a constant column to the predictor variable matrix
x_with_constant = sm.add_constant(x)

# Create and fit the linear regression model
model = sm.OLS(y, x_with_constant).fit()

# Get the predicted values and residuals
y_pred = model.predict(x_with_constant)
residuals = y - y_pred

# Breusch-Pagan test
bp_stat, bp_p_value, _, _ = het_breuschpagan(residuals, x_with_constant)
print("Breusch-Pagan test:")
print("Test Statistic:", bp_stat)
print("P-value:", bp_p_value)

# White test
white_stat, white_p_value, _, _ = het_white(residuals, x_with_constant)
print("\nWhite test:")
print("Test Statistic:", white_stat)
print("P-value:", white_p_value)
```

# Analysis of linear models

```python
# Separate the data into Lab and Field
df_lab = df[df['Location'] == 'Lab']
df_field = df[df['Location'] == 'Field']

# Linear regression for Lab
x_lab = sm.add_constant(df_lab['Post-molt'])
y_lab = df_lab['Pre-molt']
model_lab = sm.OLS(y_lab, x_lab).fit()

# Linear regression for Field
x_field = sm.add_constant(df_field['Post-molt'])
y_field = df_field['Pre-molt']
model_field = sm.OLS(y_field, x_field).fit()
```

```python
# Display model summaries
print("Linear Regression Model Summary for Lab:")
print(model_lab.summary())

print("\nLinear Regression Model Summary for Field:")
print(model_field.summary())
```

```python
# Plot separate subplots for Lab and Field
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Scatter plot and regression line for Lab
sns.scatterplot(x='Post-molt', y='Pre-molt', data=df_lab, ax=axes[0, 0], label='Lab')
axes[0, 0].plot(df_lab['Post-molt'], model_lab.predict(x_lab), label='Lab Regression Line')
axes[0, 0].set_xlabel('Post-molt')
axes[0, 0].set_ylabel('Pre-molt')
axes[0, 0].set_title('Linear Regression Model for Lab')
axes[0, 0].legend()

# Residuals vs. Predictor for Lab
sns.scatterplot(x=df_lab['Post-molt'], y=model_lab.resid, ax=axes[0, 1])
axes[0, 1].axhline(y=0, color='red', linestyle='--', label='Zero Residuals Line')
axes[0, 1].set_xlabel('Post-molt')
axes[0, 1].set_ylabel('Residuals')
axes[0, 1].set_title('Residuals vs. Predictor for Lab')
axes[0, 1].legend()

# Scatter plot and regression line for Field
sns.scatterplot(x='Post-molt', y='Pre-molt', data=df_field, ax=axes[1, 0], label='Field')
axes[1, 0].plot(df_field['Post-molt'], model_field.predict(x_field), label='Field Regression Line')
axes[1, 0].set_xlabel('Post-molt')
axes[1, 0].set_ylabel('Pre-molt')
axes[1, 0].set_title('Linear Regression Model for Field')
axes[1, 0].legend()

# Residuals vs. Predictor for Field
sns.scatterplot(x=df_field['Post-molt'], y=model_field.resid, ax=axes[1, 1])
axes[1, 1].axhline(y=0, color='red', linestyle='--', label='Zero Residuals Line')
axes[1, 1].set_xlabel('Post-molt')
axes[1, 1].set_ylabel('Residuals')
axes[1, 1].set_title('Residuals vs. Predictor for Field')
axes[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python
# Breusch-Pagan test for Lab
bp_stat_lab, bp_p_value_lab, _, _ = het_breuschpagan(model_lab.resid, x_lab)
print("Breusch-Pagan test for Lab:")
print("Test Statistic:", bp_stat_lab)
print("P-value:", bp_p_value_lab)

# White test for Lab
white_stat_lab, white_p_value_lab, _, _ = het_white(model_lab.resid, x_lab)
print("\nWhite test for Lab:")
print("Test Statistic:", white_stat_lab)
print("P-value:", white_p_value_lab)

# Breusch-Pagan test for Field
bp_stat_field, bp_p_value_field, _, _ = het_breuschpagan(model_field.resid, x_field)
print("\nBreusch-Pagan test for Field:")
print("Test Statistic:", bp_stat_field)
print("P-value:", bp_p_value_field)

# White test for Field
white_stat_field, white_p_value_field, _, _ = het_white(model_field.resid, x_field)
print("\nWhite test for Field:")
print("Test Statistic:", white_stat_field)
print("P-value:", white_p_value_field)
```
