## Appendix C: DATA AND CODE 

#### Here, In this appendix, we document the Python code for “pre-molt” and “post-molt”


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm

# Load data from CSV file
data = pd.read_csv('Crab Molt Data.csv')

df_cleaned = data.dropna()
df_cleaned.describe()

# Separate data for pre_molt and post_molt
pre_molt = df_cleaned['Pre-molt'].sort_values().to_numpy()
post_molt = df_cleaned['Post-molt'].sort_values().to_numpy()

# Descriptive statistics for post-molt variable
post_molt_max = np.max(post_molt)
post_molt_min = np.min(post_molt)
post_molt_median = np.median(post_molt)
post_molt_mean = np.mean(post_molt)
post_molt_std = np.std(post_molt)
post_molt_skewness = stats.skew(post_molt)
post_molt_kurtosis = stats.kurtosis(post_molt)

# Descriptive statistics for pre-molt variable
pre_molt_max = np.max(pre_molt)
pre_molt_min = np.min(pre_molt)
pre_molt_median = np.median(pre_molt)
pre_molt_mean = np.mean(pre_molt)
pre_molt_std = np.std(pre_molt)
pre_molt_skewness = stats.skew(pre_molt)
pre_molt_kurtosis = stats.kurtosis(pre_molt)

# Printing descriptive statistics
print("Descriptive Statistics for Post-Molt Variable:")
print("Max:", post_molt_max)
print("Min:", post_molt_min)
print("Median:", post_molt_median)
print("Mean:", post_molt_mean)
print("Standard Deviation:", post_molt_std)
print("Skewness:", post_molt_skewness)
print("Kurtosis:", post_molt_kurtosis)

print("\nDescriptive Statistics for Pre-Molt Variable:")
print("Max:", pre_molt_max)
print("Min:", pre_molt_min)
print("Median:", pre_molt_median)
print("Mean:", pre_molt_mean)
print("Standard Deviation:", pre_molt_std)
print("Skewness:", pre_molt_skewness)
print("Kurtosis:", pre_molt_kurtosis)

# Quantile plot for normality
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
stats.probplot(post_molt, dist="norm", plot=plt)
plt.title("Quantile Plot for Post-Molt Variable")

plt.subplot(1, 2, 2)
stats.probplot(pre_molt, dist="norm", plot=plt)
plt.title("Quantile Plot for Pre-Molt Variable")

plt.show()
```

### Tests for Normality

```python
# Testing for normality
print("\nTests for Normality:")
alpha = 0.05

# Anderson-Darling test
ad_stat_post, ad_crit_values_post, ad_sig_level_post = stats.anderson(post_molt, dist='norm')
ad_stat_pre, ad_crit_values_pre, ad_sig_level_pre = stats.anderson(pre_molt, dist='norm')
print("Anderson-Darling Test for Post-Molt Variable:")
print("Statistic:", ad_stat_post)
print("Critical Values:", ad_crit_values_post)
print("Significance Level:", ad_sig_level_post)
print("Is Normal (Post-Molt)?", ad_stat_post < ad_crit_values_post[2])

print("\nAnderson-Darling Test for Pre-Molt Variable:")
print("Statistic:", ad_stat_pre)
print("Critical Values:", ad_crit_values_pre)
print("Significance Level:", ad_sig_level_pre)
print("Is Normal (Pre-Molt)?", ad_stat_pre < ad_crit_values_pre[2])

# Shapiro-Wilk test
shapiro_stat_post, shapiro_p_post = stats.shapiro(post_molt)
shapiro_stat_pre, shapiro_p_pre = stats.shapiro(pre_molt)
print("\nShapiro-Wilk Test for Post-Molt Variable:")
print("Statistic:", shapiro_stat_post)
print("p-value:", shapiro_p_post)
print("Is Normal (Post-Molt)?", shapiro_p_post > alpha)

print("\nShapiro-Wilk Test for Pre-Molt Variable:")
print("Statistic:", shapiro_stat_pre)
print("p-value:", shapiro_p_pre)
print("Is Normal (Pre-Molt)?", shapiro_p_pre > alpha)

# Kolmogorov-Smirnov test
ks_stat_post, ks_p_post = stats.kstest(post_molt, 'norm')
ks_stat_pre, ks_p_pre = stats.kstest(pre_molt, 'norm')
print("\nKolmogorov-Smirnov Test for Post-Molt Variable:")
print("Statistic:", ks_stat_post)
print("p-value:", ks_p_post)
print("Is Normal (Post-Molt)?", ks_p_post > alpha)

print("\nKolmogorov-Smirnov Test for Pre-Molt Variable:")
print("Statistic:", ks_stat_pre)
print("p-value:", ks_p_pre)
print("Is Normal (Pre-Molt)?", ks_p_pre > alpha)

# Cramér-von Mises test
cm_pre = sm.stats.diagnostic.normal_ad(pre_molt)
cm_post = sm.stats.diagnostic.normal_ad(post_molt)

cm_pre_stat, cm_pre_molt_pvalue = cm_pre
cm_post_stat, cm_post_molt_pvalue = cm_post

print("\nCramér-von Mises Test for Post-Molt Variable:")
print("Statistic:", cm_pre_stat)
print("P-value:", cm_pre_molt_pvalue)
print("Is Normal (Post-Molt)?", cm_post_molt_pvalue > alpha)

print("\nCramér-von Mises Test for Pre-Molt Variable:")
print("Statistic:", cm_post_stat)
print("P-value:", cm_post_molt_pvalue)
print("Is Normal (Pre-Molt)?", cm_pre_molt_pvalue > alpha)
```

### Analysis of Normality Tests for Post-Molt and Pre-Molt Variables

The tests conducted for both the Post-Molt and Pre-Molt variables indicate that neither variable follows a normal distribution. Let's delve into the significance of the test results:

#### Anderson-Darling Test:
- The Anderson-Darling test assesses whether a sample of data comes from a specific distribution, in this case, the normal distribution.
- For both the Post-Molt and Pre-Molt variables, the statistic values exceed the critical values at all significance levels (15%, 10%, 5%, 2.5%, and 1%), indicating a rejection of the null hypothesis of normality.
- Therefore, we conclude that neither the Post-Molt nor the Pre-Molt variable follows a normal distribution based on the Anderson-Darling test.

#### Shapiro-Wilk Test:
- The Shapiro-Wilk test also evaluates the normality of a dataset.
- For both variables, the p-values obtained from the Shapiro-Wilk test are significantly less than the conventional significance level of 0.05, indicating strong evidence against the null hypothesis of normality.
- Thus, we reject the assumption of normality for both the Post-Molt and Pre-Molt variables based on the Shapiro-Wilk test.

#### Kolmogorov-Smirnov Test:
- The Kolmogorov-Smirnov test compares the empirical distribution function of the data with a reference cumulative distribution function, often the normal distribution in this context.
- In both cases, the p-values are extremely small (close to 0), leading to the rejection of the null hypothesis of normality.
- Consequently, neither the Post-Molt nor the Pre-Molt variable can be considered normally distributed based on the Kolmogorov-Smirnov test.

#### Cramér-von Mises Test:
- The Cramér-von Mises test is another statistical test used to assess the goodness of fit of a sample to a given distribution.
- For both variables, the p-values are extremely small, providing strong evidence against the null hypothesis of normality.
- Therefore, we reject the assumption of normality for both the Post-Molt and Pre-Molt variables based on the Cramér-von Mises test.

In summary, the results of these tests suggest that neither the Post-Molt nor the Pre-Molt variable follows a normal distribution. This has implications for statistical analyses and modeling, as methods assuming normality may not be appropriate for these variables.

### Linear Model using Scatter Plot:

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd

# Load data from CSV file
data = pd.read_csv('Crab Molt Data.csv')

# Drop rows with missing values
df_cleaned = data.dropna()

# Extract pre-molt and post-molt sizes
x = df_cleaned['Post-molt'].to_numpy().reshape(-1, 1)
y = df_cleaned['Pre-molt'].to_numpy()

# Fit the linear regression model using scikit-learn
model = LinearRegression().fit(x, y)

# Extract the intercept and slope from the model
b = model.intercept_
m = model.coef_[0]

plt.figure(figsize=(8, 6))
# Plot the data points
plt.plot(x, y, 'o', label='Data',alpha=0.5, color='blue')

# Add the linear regression line to the scatterplot
plt.plot(x, m*x + b, color='red', label='Linear Regression')

plt.xlabel("Post molt")
plt.ylabel("Pre molt")
plt.title("Linear Model")
plt.legend()
plt.show()

```

### Residual Plot:

```python
# Calculate Pearson's R-squared value
R_Square = model.score(x, y)
print("Coeff of Determination: ", R_Square)


# Compute residuals
y_pred = model.predict(x)
residuals = y - y_pred


# Plot the residuals
plt.figure(figsize=(8, 6))
plt.scatter(y_pred, residuals, color='blue', alpha=0.5)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residual Plot')
plt.xlabel('Pre-Molt Sizes')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()

```

### Quatile-Quantile Plot:

```python
# Check for normality of residuals
skewness = stats.skew(residuals)
kurtosis = stats.kurtosis(residuals)
plt.figure(figsize=(12, 6))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title('Quantile Plot of Residuals')
plt.show()
```

### Checks for Normality and Heteroskedasticity:

```python
# Perform normality tests
ks_statistic, ks_p_value = stats.kstest(residuals, 'norm')
ad_statistic, ad_critical_values, _ = stats.anderson(residuals, dist='norm')
shapiro_statistic, shapiro_p_value = stats.shapiro(residuals)

# Check for heteroskedasticity
# Visual estimation from the plot of residuals can be performed above
# Breusch–Pagan test
X = sm.add_constant(x) 
bp_test = sm.stats.diagnostic.het_breuschpagan(residuals, X)
white_test = sm.stats.diagnostic.het_white(residuals, X)

# Print the results
rs = pd.DataFrame(residuals, columns=['Residuals'])
rs.describe()
print("Pearson's R-squared Value:", R_Square)
print("Skewness of Residuals:", skewness)
print("Kurtosis of Residuals:", kurtosis)
print("Kolmogorov-Smirnov Test p-value:", ks_p_value)
print("Anderson-Darling Test Statistic:", ad_statistic)
print("Shapiro-Wilk Test p-value:", shapiro_p_value)
print("Breusch–Pagan Test p-value:", bp_test[1])
print("White Test p-value:", white_test[1])
```

### Checks for Normality and Heteroskedasticity For Lab Versus Field Crabs :

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import scipy.stats as stats
import matplotlib.pyplot as plt

# Load data from CSV file
data = pd.read_csv('Crab Molt Data.csv')
# Drop rows with missing values
df_cleaned = data.dropna()
# Get unique locations
locations = df_cleaned['Location'].unique()
# Create scatterplots for each location
for location in locations:
    # Filter data for the current location
    subset = df_cleaned[df_cleaned['Location'] == location]
    # Extract pre-molt and post-molt sizes for the current location
    pre_molt_sizes = subset['Pre-molt']
    post_molt_sizes = subset['Post-molt']
    # Fit the linear regression model
    X = sm.add_constant(post_molt_sizes)  # Add a constant term to the predictor
    model = sm.OLS(pre_molt_sizes, X)
    results = model.fit()
    # Calculate Pearson's R-squared value
    r_squared = results.rsquared
    # Compute residuals
    residuals = results.resid
    # Plot the residuals and quantile plot side by side
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))
    # Residual plot
    axes[0].scatter(pre_molt_sizes, residuals, color='blue', alpha=0.5)
    axes[0].axhline(y=0, color='red', linestyle='--')
    axes[0].set_title(f'{location} Residual Plot')
    axes[0].set_xlabel('Pre-Molt Sizes')
    axes[0].set_ylabel('Residuals')
    axes[0].grid(True)
    # Quantile plot of residuals
    stats.probplot(residuals, dist="norm", plot=axes[1])
    axes[1].set_title(f'{location} Quantile Plot of Residuals')
    plt.tight_layout()
    plt.show()
    # Check for normality of residuals
    skewness = stats.skew(residuals)
    kurtosis = stats.kurtosis(residuals)

    # Perform normality tests
    ks_statistic, ks_p_value = stats.kstest(residuals, 'norm')
    ad_statistic, ad_critical_values, _ = stats.anderson(residuals, dist='norm')
    shapiro_statistic, shapiro_p_value = stats.shapiro(residuals)

    # Check for heteroskedasticity
    # Breusch–Pagan test
    bp_test = sm.stats.diagnostic.het_breuschpagan(residuals, X)
    # White test
    white_test = sm.stats.diagnostic.het_white(residuals, X)

    # Print the results for each location
    print(f"\nResults for Location: {location}")
    print("Pearson's R-squared Value:", r_squared)
    print("Skewness of Residuals:", skewness)
    print("Kurtosis of Residuals:", kurtosis)
    print("Kolmogorov-Smirnov Test p-value:", ks_p_value)
    print("Anderson-Darling Test Statistic:", ad_statistic)
    print("Shapiro-Wilk Test p-value:", shapiro_p_value)
    print("Breusch–Pagan Test p-value:", bp_test[1])
    print("White Test p-value:", white_test[1])
```

### K-Fold Cross Validation To Estimate Prediction Accuracy of Linear Models:

```python
# Load data from CSV file
data = pd.read_csv('Crab Molt Data.csv')

# Drop rows with missing values
df_cleaned = data.dropna()

# Extract pre-molt and post-molt sizes
pre_molt_sizes = df_cleaned['Pre-molt']
post_molt_sizes = df_cleaned['Post-molt']

# Define the number of folds for cross-validation
k = 5
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# Initialize lists to store evaluation metrics
r_squared_values = []
mse_values = []

X = sm.add_constant(post_molt_sizes)  # Add a constant term to the predictor

# Perform K-fold cross-validation
for i, (train_index, test_index) in enumerate(kf.split(X)):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = pre_molt_sizes.iloc[train_index], pre_molt_sizes.iloc[test_index]

    # Fit the linear regression model
    model = sm.OLS(y_train, X_train)
    results = model.fit()

    # Make predictions on the test set
    y_pred = results.predict(X_test)

    # Compute evaluation metrics
    r_squared = results.rsquared
    mse = np.mean((y_test - y_pred) ** 2)

    # Store evaluation metrics
    r_squared_values.append(r_squared)
    mse_values.append(mse)

    # Print evaluation metrics for each fold
    print(f"Fold {i+1}:")
    print(f"R-squared: {r_squared}")
    print(f"MSE: {mse}")
    print("-" * 30)

# Print average evaluation metrics across all folds
print("Average across all folds:")
print(f"Average R-squared: {np.mean(r_squared_values)}")
print(f"Average MSE: {np.mean(mse_values)}")
```

#### The K-fold cross-validation process provides valuable insights into the performance of our models. Here's a summary of the average R-squared values and mean squared error (MSE) across the folds:

- **Average R-squared Values**: The R-squared value, also known as the coefficient of determination, measures the proportion of the variance in the dependent variable that is predictable from the independent variables. In our case, the average R-squared values range from approximately 0.9809 to 0.9838 across the folds. These high R-squared values indicate that our models explain a significant amount of the variance in the data, suggesting good predictive capability.
  
- **Average MSE Values**: The mean squared error (MSE) quantifies the average squared difference between the actual and predicted values. Lower MSE values indicate better model performance. Across the folds, the average MSE values range from approximately 5.0337 to 8.2271. These relatively low MSE values suggest that our models have reasonably accurate predictions.

##### Overall, the K-fold cross-validation results indicate that our models perform well in predicting pre-molt crab sizes based on post-molt sizes. The high R-squared values and low MSE values across the folds provide confidence in the predictive capability and accuracy of our models.
