```python
import pandas as pd

# load data into a pandas dataframe
df = pd.read_csv('Preliminary college year.csv')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python

# create binary indicator variables for categorical features
#df = pd.get_dummies(df, columns=['Gender', 'Federal Ethnic Group'])
df = df.drop(['Gender', 'Federal Ethnic Group'], axis=1)

df.describe()
```

### Fill NaN values with median

```python
df.fillna(df.median(numeric_only=True).round(1), inplace=True)
df.head()
```

### Drop the 'Reason for not Completing Connect' and 'Reason not Retained' are different for different person, so drop it for some time so that we can evaluate

```python
df = df.drop('Reason for not Completing Connect', axis=1)
df = df.drop('Reason not Retained', axis=1)
```

```python
df.head()
```

### Split the data into training and testing sets to evaluate the performance of the logistic model.

```python
import pandas as pd

df['Completed Connect? (1=yes, 0=no)'] = df['Completed Connect? (1=yes, 0=no)'].replace({'contract for fall': 0})

# Convert the column to numeric type
df['Completed Connect? (1=yes, 0=no)'] = pd.to_numeric(df['Completed Connect? (1=yes, 0=no)'], errors='coerce')

# Drop rows with NaN values
df.dropna(inplace=True)

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop('Retained F17-F18? (1=yes, 0=no)', axis=1), df['Retained F17-F18? (1=yes, 0=no)'].astype(int), test_size=0.2, random_state=102)
```

```python
from sklearn.model_selection import train_test_split

# split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop('Retained F17-F18', axis=1), df['Retained F17-F18'].astype(int), test_size=0.2, random_state=102)
```

### Fit a logistic regression model to the training data, using a subset of the variables that are most likely to be significant predictors of the outcome based on domain knowledge or exploratory analysis.

```python
# Initialize and fit a logistic regression model on the training data
model = LogisticRegression(max_iter=8000, solver='sag')
model.fit(X_train, y_train)
```

### Evaluate the performance of the model


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f'Accuracy: {accuracy:.3f}')
print(f'Precision: {precision:.3f}')
print(f'Recall: {recall:.3f}')
print(f'F1 score: {f1:.3f}')
```

### Coefficient values: In logistic regression, the coefficients represent the amount of change in the log odds of the outcome variable for a unit change in the predictor variable. Larger absolute values of the coefficients indicate greater importance of the corresponding predictor variable.

```python
X = df.iloc[:,0:28]

df['Retained F17-F18? (1=yes, 0=no)'] = df['Retained F17-F18? (1=yes, 0=no)'].astype(int)
y = df['Retained F17-F18? (1=yes, 0=no)']

# Identify important features
coefficients = pd.DataFrame({"feature": X.columns, "coefficient": model.coef_[0]})
coefficients = coefficients.sort_values(by="coefficient", ascending=False)
print("Important features:\n", coefficients)
```

### Diagnostic plots: Residuals vs. Fitted Values, Q-Q plot of Residuals, and Residuals vs. Leverage plot

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Calculate residuals
y_pred = model.predict(X)
residuals = y - y_pred

# Residuals vs. Fitted values plot
plt.figure(figsize=(8, 6))
sns.residplot(x=y_pred, y=residuals, lowess=True, line_kws={'color': 'red', 'lw': 1})
plt.title('Residuals vs. Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()

# Q-Q plot of residuals
plt.figure(figsize=(8, 6))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title('Q-Q plot of Residuals')
plt.show()

# Calculate leverage
n = X.shape[0]
p = X.shape[1]
hat_matrix = np.matmul(X, np.linalg.solve(np.matmul(X.T, X), X.T))
leverage = np.diag(hat_matrix)

# Leverage plot
plt.figure(figsize=(8, 6))
sns.regplot(x=leverage, y=residuals, lowess=True, line_kws={'color': 'red', 'lw': 1})
plt.title('Residuals vs. Leverage')
plt.xlabel('Leverage')
plt.ylabel('Residuals')
plt.show()
```

### The precision-recall curve and find the optimal threshold for maximizing the F1 score.

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# Calculate precision and recall for different thresholds
precision, recall, thresholds = precision_recall_curve(y, model.predict_proba(X)[:,1])

# Plot precision-recall curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', label='Logistic Regression')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.show()

# Find optimal threshold for specific metric (e.g., F1 score)
f1_scores = 2 * (precision * recall) / (precision + recall)
optimal_threshold = thresholds[np.argmax(f1_scores)]
print("Optimal Threshold (F1 Score):", optimal_threshold)
```
