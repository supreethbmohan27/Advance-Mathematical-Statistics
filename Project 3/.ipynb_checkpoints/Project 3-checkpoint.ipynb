```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load data from CSV file into a DataFrame
data = pd.read_csv('IndiaPoliticalViolence(no Islands_Kashmir_Ladakh).csv')    

# Select only the latitude and longitude columns for clustering
coordinates = data[['latitude', 'longitude']]

# Convert coordinates to a numpy array
coordinates_array = coordinates.values

# Specify the number of clusters (you need to decide this based on your data)
num_clusters = 2

# Perform K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
kmeans.fit(coordinates_array)

# Get cluster centers
cluster_centers = kmeans.cluster_centers_

# Get cluster labels for each data point
labels = kmeans.labels_

# Plot the clustered data points
plt.figure(figsize=(10, 8))  # Reverse the figure size

# Plot each cluster
for i in range(num_clusters):
    cluster_points = coordinates_array[labels == i]
    plt.scatter(cluster_points[:, 1], cluster_points[:, 0], label=f'Cluster {i+1}')  # Invert x and y

# Plot cluster centers
plt.scatter(cluster_centers[:, 1], cluster_centers[:, 0], color='black', marker='x', label='Cluster Center')  # Invert x and y

plt.ylabel('Latitude')  # Invert x and y labels
plt.xlabel('Longitude')  # Invert x and y labels
plt.title('K-means Clustering of Political Violence Data')
plt.legend()
plt.grid(True)

# Rotate the plot
plt.gca().set_aspect('equal', adjustable='box')  # Adjust aspect ratio

plt.show()


```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import folium

# Load data from CSV file into a DataFrame
data = pd.read_csv('IndiaPoliticalViolence(no Islands_Kashmir_Ladakh).csv')    

# Select only the latitude and longitude columns for clustering
coordinates = data[['latitude', 'longitude']]

# Convert coordinates to a numpy array
coordinates_array = coordinates.values

# Specify the number of clusters (you need to decide this based on your data)
num_clusters = 5

# Perform K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
kmeans.fit(coordinates_array)

# Get cluster centers
cluster_centers = kmeans.cluster_centers_

# Create a map centered around India
india_map = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

# Add clustered data points to the map
for i in range(len(coordinates_array)):
    folium.CircleMarker(
        location=[coordinates_array[i][0], coordinates_array[i][1]],
        radius=5,
        color=f'blue' if kmeans.labels_[i] == 0 else ('green' if kmeans.labels_[i] == 1 else ('red' if kmeans.labels_[i] == 2 else ('purple' if kmeans.labels_[i] == 3 else 'orange'))),
        fill=True,
        fill_color=f'blue' if kmeans.labels_[i] == 0 else ('green' if kmeans.labels_[i] == 1 else ('red' if kmeans.labels_[i] == 2 else ('purple' if kmeans.labels_[i] == 3 else 'orange'))),
    ).add_to(india_map)

# Add cluster centers to the map
for center in cluster_centers:
    folium.Marker(
        location=[center[0], center[1]],
        icon=folium.Icon(color='black', icon='info-sign')
    ).add_to(india_map)

# Display the map
india_map.save('india_clusters_map.html')
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import folium

# Load data from CSV file into a DataFrame
data = pd.read_csv('IndiaPoliticalViolence(no Islands_Kashmir_Ladakh).csv')    

# Select only the latitude and longitude columns for clustering
coordinates = data[['latitude', 'longitude']]

# Convert coordinates to a numpy array
coordinates_array = coordinates.values

# Specify the number of clusters
num_clusters = 5

# Perform K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
kmeans.fit(coordinates_array)

# Calculate the number of violent incidents in each cluster
cluster_violence_counts = np.zeros(num_clusters)

for i in range(len(coordinates_array)):
    cluster_label = kmeans.labels_[i]
    cluster_violence_counts[cluster_label] += 1

# Create a dictionary to store cluster labels and their corresponding violence counts
cluster_counts_dict = {label: count for label, count in enumerate(cluster_violence_counts)}

# Sort the clusters based on violence counts
sorted_clusters = sorted(cluster_counts_dict.items(), key=lambda x: x[1])

# Create a map centered around India
india_map = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

# Visualize the clusters based on violence counts
for cluster_label, count in sorted_clusters:
    cluster_indices = np.where(kmeans.labels_ == cluster_label)[0]
    for i in cluster_indices:
        if cluster_label == sorted_clusters[0][0]:
            color = 'green'  # Least violent cluster
        elif cluster_label == sorted_clusters[-1][0]:
            color = 'red'  # Most violent cluster
        else:
            color = 'blue'  # Clusters in between
        folium.CircleMarker(
            location=[coordinates_array[i][0], coordinates_array[i][1]],
            radius=5,
            color=color,
            fill=True,
            fill_color=color,
        ).add_to(india_map)

# Add cluster centers to the map
for center in kmeans.cluster_centers_:
    folium.Marker(
        location=[center[0], center[1]],
        icon=folium.Icon(color='black', icon='info-sign')
    ).add_to(india_map)

# # Retrieve detailed information for the most violent cluster
# most_violent_cluster_data = data[kmeans.labels_ == most_violent_cluster_label]

# # Retrieve detailed information for the least violent cluster
# least_violent_cluster_data = data[kmeans.labels_ == least_violent_cluster_label]

# # Retrieve detailed information for clusters in between
# in_between_clusters_data = {}
# for cluster_label, _ in sorted_clusters[1:-1]:
#     in_between_clusters_data[cluster_label] = data[kmeans.labels_ == cluster_label]

# # Print information for the most violent cluster
# print("Most Violent Cluster:")
# print(most_violent_cluster_data)

# # Print information for the least violent cluster
# print("\nLeast Violent Cluster:")
# print(least_violent_cluster_data)

# # Print information for clusters in between
# print("\nClusters In Between:")
# for cluster_label, cluster_data in in_between_clusters_data.items():
#     print(f"\nCluster {cluster_label}:")
#     print(cluster_data)

# Print the most violent cluster
most_violent_cluster_label, most_violent_count = sorted_clusters[-1]
print(f"The most violent cluster is Cluster {most_violent_cluster_label} with {most_violent_count} incidents.")

# Print the least violent cluster
least_violent_cluster_label, least_violent_count = sorted_clusters[0]
print(f"The least violent cluster is Cluster {least_violent_cluster_label} with {least_violent_count} incidents.")

# Print clusters in between
for cluster_label, count in sorted_clusters[1:-1]:
    print(f"Cluster {cluster_label} has {count} incidents.")

    
# Save the map
india_map.save('india_clusters_map_with_violence_counts.html')
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import folium

# Load data from CSV file into a DataFrame
data = pd.read_csv('IndiaPoliticalViolence(no Islands_Kashmir_Ladakh).csv')    

# Select only the latitude and longitude columns for clustering
coordinates = data[['latitude', 'longitude']]

# Convert coordinates to a numpy array
coordinates_array = coordinates.values

# Specify the number of clusters
num_clusters = 5

# Perform K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
kmeans.fit(coordinates_array)

# Calculate the number of violent incidents in each cluster
cluster_violence_counts = np.zeros(num_clusters)

for i in range(len(coordinates_array)):
    cluster_label = kmeans.labels_[i]
    cluster_violence_counts[cluster_label] += 1

# Create a dictionary to store cluster labels and their corresponding violence counts
cluster_counts_dict = {label: count for label, count in enumerate(cluster_violence_counts)}

# Sort the clusters based on violence counts
sorted_clusters = sorted(cluster_counts_dict.items(), key=lambda x: x[1])

# Define a color palette for clusters
color_palette = ['green', 'blue', 'red', 'orange', 'purple']  # You can add more colors if needed

# Create a dictionary mapping cluster labels to colors
cluster_colors = {cluster_label: color_palette[i % len(color_palette)] for i, (cluster_label, _) in enumerate(sorted_clusters)}

# Create a map centered around India
india_map = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

# Visualize the clusters based on violence counts
for cluster_label, count in sorted_clusters:
    cluster_indices = np.where(kmeans.labels_ == cluster_label)[0]
    color = cluster_colors[cluster_label]
    for i in cluster_indices:
        folium.CircleMarker(
            location=[coordinates_array[i][0], coordinates_array[i][1]],
            radius=5,
            color=color,
            fill=True,
            fill_color=color,
        ).add_to(india_map)

# Add cluster centers to the map
for center, color, count in zip(kmeans.cluster_centers_, color_palette, cluster_violence_counts):
    folium.Marker(
        location=[center[0], center[1]],
        icon=folium.Icon(color='black', icon='info-sign')
    ).add_to(india_map)
    # Add label indicating incident count
    folium.Marker(
        location=[center[0], center[1]],
        icon=folium.DivIcon(html=f"<div style='font-size: 12pt; color: black;'>{count}</div>")
    ).add_to(india_map)


# Save the map
india_map.save('india_clusters_map_with_violence_counts.html')

# Print the colors of the clusters along with their names and incident counts
print("Cluster Colors and Incidents:")
for cluster_label, count in sorted_clusters:
    color = cluster_colors[cluster_label]
    print(f"Cluster {cluster_label} (Color: {color}) has {count} incidents.")

# Print the most violent cluster
most_violent_cluster_label, most_violent_count = sorted_clusters[-1]
print(f"\nThe most violent cluster is Cluster {most_violent_cluster_label} with {most_violent_count} incidents.")

# Print the least violent cluster
least_violent_cluster_label, least_violent_count = sorted_clusters[0]
print(f"The least violent cluster is Cluster {least_violent_cluster_label} with {least_violent_count} incidents.")

# # Print clusters in between
# for cluster_label, count in sorted_clusters[1:-1]:
#     print(f"Cluster {cluster_label} has {count} incidents.")
    
# Print the clusters and their counts in descending order
sorted_clusters_descending = sorted(sorted_clusters, key=lambda x: x[1], reverse=True)

# Print the clusters and their counts
print("Clusters and Incidents (Descending Order):")
for cluster_label, count in sorted_clusters_descending:
    print(f"Cluster {cluster_label} has {count} incidents.")

    
# Assuming 'latitude' and 'longitude' are the column names in your DataFrame
top_repeated_latitudes = data['latitude'].value_counts().head(5)
top_repeated_longitudes = data['longitude'].value_counts().head(5)

print("Top 5 Repeated Latitudes:")
print(top_repeated_latitudes)

print("\nTop 5 Repeated Longitudes:")
print(top_repeated_longitudes)
    
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import folium

# Load data from CSV file into a DataFrame
data = pd.read_csv('IndiaPoliticalViolence(no Islands_Kashmir_Ladakh).csv')

# Select relevant columns for analysis
relevant_data = data[['fatalities', 'event_type', 'latitude', 'longitude']]

# Group data by event type and sum up fatalities for each event type
event_fatalities = relevant_data.groupby('event_type')['fatalities'].sum().reset_index()

# Sort event_fatalities based on fatalities in descending order and select top 5
top_5_fatalities_events = event_fatalities.sort_values(by='fatalities', ascending=False).head(5)

# Specify the number of clusters
num_clusters = 6

# Select only the latitude, longitude, and fatalities columns for clustering
coordinates = relevant_data[['latitude', 'longitude', 'fatalities']]

# Convert coordinates to a numpy array
coordinates_array = coordinates.values

# Perform K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
kmeans.fit(coordinates_array)

# Define cluster colors
cluster_colors = {0: 'blue', 1: 'green', 2: 'red', 3: 'orange', 4: 'purple', 5: 'yellow'}  # Define colors for each cluster label

# Initialize a dictionary to store data for each cluster
cluster_data = {label: relevant_data[kmeans.labels_ == label] for label in range(num_clusters)}

# Print data for each cluster
for label in range(num_clusters):
    print(f"\nCluster {label + 1}: {cluster_colors[label]}")
    print("Top 5 event types causing the most fatalities:")
    print(cluster_data[label].groupby('event_type')['fatalities'].sum().reset_index().nlargest(5, 'fatalities'))

# Create a map centered around India
india_map = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

# Visualize the clusters based on fatalities
for i in range(len(coordinates_array)):
    cluster_label = kmeans.labels_[i]
    color = cluster_colors[cluster_label]
    folium.CircleMarker(
        location=[coordinates_array[i][0], coordinates_array[i][1]],
        radius=5,
        color=color,
        fill=True,
        fill_color=color,
    ).add_to(india_map)

# Add cluster centers to the map and print cluster details
for center, cluster_label in zip(kmeans.cluster_centers_, kmeans.labels_):
    color = cluster_colors[cluster_label]
    print(f"Cluster {cluster_label + 1}: {color}")
    folium.Marker(
        location=[center[0], center[1]],
        icon=folium.Icon(color='black', icon='info-sign')
    ).add_to(india_map)

# Save the map
india_map.save('india_clusters_map_with_fatalities.html')
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import folium

# Load data from CSV file into a DataFrame
data = pd.read_csv('IndiaPoliticalViolence(no Islands_Kashmir_Ladakh).csv')

# Select relevant columns for analysis
relevant_data = data[['fatalities', 'event_type', 'latitude', 'longitude']]

# Group data by event type and sum up fatalities for each event type
event_fatalities = relevant_data.groupby('event_type')['fatalities'].sum().reset_index()

# Sort event_fatalities based on fatalities in descending order and select top 5
top_5_fatalities_events = event_fatalities.sort_values(by='fatalities', ascending=False).head(5)

# Specify the number of clusters
num_clusters = 6

# Select only the latitude, longitude, and fatalities columns for clustering
coordinates = relevant_data[['latitude', 'longitude', 'fatalities']]

# Convert coordinates to a numpy array
coordinates_array = coordinates.values

# Perform K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
kmeans.fit(coordinates_array)

# Define cluster colors
cluster_colors = {0: 'blue', 1: 'green', 2: 'red', 3: 'orange', 4: 'purple', 5: 'yellow'}  # Define colors for each cluster label

# Print data for each cluster and create map
india_map = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

for label in range(num_clusters):
    # Initialize a dictionary to store data for each cluster
    cluster_data = relevant_data[kmeans.labels_ == label]

    print(f"\nCluster {label + 1}: {cluster_colors[label]}")
    print("Top 5 event types causing the most fatalities:")
    print(cluster_data.groupby('event_type')['fatalities'].sum().nlargest(5).reset_index())

    # Visualize the clusters based on fatalities and add cluster details as labels
    for i in range(len(coordinates_array)):
        cluster_label = kmeans.labels_[i]
        color = cluster_colors[cluster_label]
        folium.CircleMarker(
            location=[coordinates_array[i][0], coordinates_array[i][1]],
            radius=5,
            color=color,
            fill=True,
            fill_color=color,
        ).add_to(india_map)

    # Add cluster centers to the map and print cluster details
    center = kmeans.cluster_centers_[label]
    color = cluster_colors[label]
    top_events_html = ''
    top_events = cluster_data.groupby('event_type')['fatalities'].sum().nlargest(5).reset_index()
    top_events_html = '<br>'.join([f"{row['event_type']}: {row['fatalities']}" for _, row in top_events.iterrows()])
    folium.Marker(
        location=[center[0], center[1]],
        icon=folium.DivIcon(html=f"<div style='color: black; font-size: 12pt; padding: 5px;'>Cluster: {label + 1}<br>Fatalities: {center[2]}<br>Top Events:<br>{top_events_html}</div>"),
    ).add_to(india_map)

# Save the map
india_map.save('india_clusters_map_with_fatalities.html')
```
